```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

def scale_features(df, columns):
    scaled = None
### WRITE SOLUTION HERE
    scaler = StandardScaler()
    df_scaled = df.copy()
    df_scaled[columns] = scaler.fit_transform(df[columns])
    scaled = df_scaled
    return scaled
```

```python
# Tests 10 points.
df = pd.DataFrame({"a":[1,2,3], "b":[10,10,10]})
res = scale_features(df, ["a"])

# column "a" must be standardized (mean ~0, std ~1)
assert abs(res["a"].mean()) < 1e-6
assert abs(res["a"].std() - 1) < 1e-6
# column "b" unchanged
assert all(res["b"] == 10)
```
