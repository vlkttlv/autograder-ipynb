### Проверка на четность

```python
def is_even(n):
    result = False
    ### BEGIN SOLUTION
    result = n % 2 == 0
    ### END SOLUTION
    return result
```

```python
even_check = is_even(10)
print(f"Is even = {even_check}")
```

```python
# Tests 3 points.
assert is_even(1) is False
assert is_even(-2) is True
### BEGIN HIDDEN TESTS
assert is_even(2) is True
assert is_even(3) is False
### END HIDDEN TESTS
```
