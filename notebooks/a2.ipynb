```python
def calculate_average(numbers):
    res = 0
    ### BEGIN SOLUTION
    res = sum(numbers) / len(numbers)
    ### END SOLUTION
    return res
```

```python
# Tests 10 points.

assert calculate_average([10, 10, 10]) == 10
assert not calculate_average([1, 2, 3, 4, 5]) == 5

### BEGIN HIDDEN TESTS
assert calculate_average([1, 2, 3, 4, 5]) == 3
assert calculate_average([1, 1, 1, 1, 1]) == 1
### END HIDDEN TESTS
```

```python
def sum_nums(nums):
    res = 0
    ### BEGIN SOLUTION
    for num in nums:
        res += num
    ### END SOLUTION
    return res
```

```python
# Tests 5 points.

assert sum_nums([10, 10, 10]) == 30
assert not sum_nums([1, 2, 3, 4, 5]) == 80

### BEGIN HIDDEN TESTS
assert sum_nums([1, 2, 3, 4, 5]) == 15
assert sum_nums([1, 1, 1, 1, 1]) == 5
### END HIDDEN TESTS
```
