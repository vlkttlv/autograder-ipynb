### Факториал числа

```python
def factorial(n):
    result = 1
    ### BEGIN SOLUTION
    for i in range(1, n + 1):
        result *= i
    ### END SOLUTION
    return result
```

```python
f = factorial(5)
print(f"Factorial = {f}")
```

```python
# Tests 20 points.
assert factorial(5) == 120
assert factorial(3) == 6
### BEGIN HIDDEN TESTS
assert factorial(0) == 1
assert factorial(4) == 24
### END HIDDEN TESTS
```
