```python
import pandas as pd
from sklearn.model_selection import train_test_split

def make_split(df, target, test_size=0.2, random_state=42):
    parts = None
    ### BEGIN SOLUTION
    X = df.drop(columns=[target])
    y = df[target]
    parts = train_test_split(X, y, test_size=test_size, random_state=random_state)
    ### END SOLUTION
    return parts  # X_train, X_test, y_train, y_test
```

```python
# Tests 35 points.
df = pd.DataFrame({
    "f1":[1,2,3,4,5],
    "f2":[10,20,30,40,50],
    "target":[0,1,0,1,0]
})

X_train, X_test, y_train, y_test = make_split(df, "target", test_size=0.4, random_state=0)
assert X_train.shape[0] == 3
assert X_test.shape[0] == 2
assert len(y_train) == 3

### BEGIN HIDDEN TESTS
X_train2, X_test2, y_train2, y_test2 = make_split(df, "target", test_size=0.2, random_state=0)
assert X_test2.shape[0] == 1
assert len(y_train2) == 4
### END HIDDEN TESTS
```
